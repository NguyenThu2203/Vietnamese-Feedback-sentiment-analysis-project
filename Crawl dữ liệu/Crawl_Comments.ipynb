```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
```

```python
comments_dataset = []
for i in range(1, 4):
    page_url = 'https://tinhte.vn/thread/ios-18-se-chuyen-mau-icon-app-o-dark-mode-cho-phep-khoa-app-bang-face-id.3795955/page-%s' %i
    page = requests.get(page_url)
    soup = BeautifulSoup(page.content, 'html.parser')
    comments = soup.find('div', class_='jsx-1984507624 thread-comments__container').find_all('div',class_="jsx-691990575 thread-comment__container")
    for comment in comments:
        print("\n")
        content=comment.find('div',class_="jsx-691990575 thread-comment__content").find('span',class_="xf-body-paragraph").text.strip()
        print(content)
        comments_dataset.append((content,))
```

```python
comments_df = pd.DataFrame(comments_dataset, columns=['Comment'])

# Đặt tên cho tệp CSV
csv_file = 'comments_dataset.csv'

# Lưu DataFrame vào tệp CSV
if os.path.exists(csv_file):
    comments_df.to_csv(csv_file, mode='a', header=False, index=None, encoding='utf-8')
else:
    comments_df.to_csv(csv_file, index=None, encoding='utf-8')
```
